# 1. Image Filtering

```python
# Image filtering helps in reducing noise and improving the quality of the image

import cv2
from matplotlib import pyplot as plt
```

```python
image1 = cv2.imread("/content/indoor1.png")
```

```python
# Apply Gaussian Blur
blurred_image = cv2.GaussianBlur(image1, (5, 5), 0)
```

```python
# Display the original and blurred images
plt.subplot(121), plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)),
plt.title('Original Image')
plt.subplot(122), plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)),
plt.title('Blurred Image')
plt.show()
```

# 2. Edge Detection

```python
# Edge detection is useful for identifying the boundaries of objects within images.

# Convert the image to grayscale
gray_image = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)

# Apply Canny edge detection
edges = cv2.Canny(gray_image, 100, 200)

# Display the edges
plt.imshow(edges, cmap='gray')
plt.title('Edge Image')
plt.show()
```

# 3. Image Transformation

```python
# Rotate the image by 45 degrees
(h, w) = image1.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_image = cv2.warpAffine(image1, M, (w, h))

# Display the rotated image
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image')
plt.show()
```

# 4. Image Enhancement

```python
# Image enhancement techniques like histogram equalization can improve the visibility of objects in images.

# Convert the image to grayscale
gray_image = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(gray_image)

# Display the original and equalized images
plt.subplot(121), plt.imshow(gray_image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(equalized_image, cmap='gray'), plt.title('Equalized Image')
plt.show()
```

# Morphological Operations

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Morphological operations are useful for removing noise and small objects from binary images.

study_room = cv2.imread('/content/study_room.png', 0)
```

```python
# Apply binary thresholding
_, binary_image = cv2.threshold(study_room, 127, 255, cv2.THRESH_BINARY)
```

```python
# Define a kernel
kernel = np.ones((5, 5), np.uint8)
```

```python
# Apply dilation
dilated_image = cv2.dilate(binary_image, kernel, iterations=1)
# Apply erosion
eroded_image = cv2.erode(binary_image, kernel, iterations=1)
```

```python
# Display the results
plt.subplot(131), plt.imshow(binary_image, cmap='gray'), plt.title('Binary Image')
plt.subplot(132), plt.imshow(dilated_image, cmap='gray'), plt.title('Dilated Image')
plt.subplot(133), plt.imshow(eroded_image, cmap='gray'), plt.title('Eroded Image')
plt.show()
```

# Fast R-CNN

```python
# Fast R-CNN improves upon R-CNN by processing the entire image with a CNN to generate a feature map and then using region proposals to classify and localize objects.

import torch
import torchvision
from torchvision.models.detection import fasterrcnn_resnet50_fpn
```

```python
# Load a pre-trained Fast R-CNN model (using Faster R-CNN as a base)
model = fasterrcnn_resnet50_fpn(pretrained=True)
model.eval()
```

```python
# Load an image
from PIL import Image
from torchvision import transforms
```

```python
from typing import List
used_Fast_RNN_study_table = Image.open('/content/study_table.png')
used_Fast_RNN_study_room = Image.open('/content/study_room.png')
used_Fast_RNN_desi_dining_room = Image.open('/content/desi_dining_room.png')
used_Fast_RNN_cylinder_factory = Image.open('/content/cylinderFactory.png')
transform = transforms.Compose([transforms.ToTensor()])

updated_image1 = transform(used_Fast_RNN_study_table)
updated_image2 = transform(used_Fast_RNN_study_room)
updated_image3 = transform(used_Fast_RNN_desi_dining_room)
updated_image4 = transform(used_Fast_RNN_cylinder_factory)

# Perform object detection
with torch.no_grad():
    prediction1 = model([updated_image1])
    prediction2 = model([updated_image2])
    prediction3 = model([updated_image3])
    prediction4 = model([updated_image4])

# Print the results
print(prediction1)
print(prediction2)
print(prediction3)
print(prediction4)
```
